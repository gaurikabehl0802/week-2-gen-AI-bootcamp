```python
# 1. Install required libraries
!pip install -q google-genai gradio

import gradio as gr
from google import genai
from google.genai import types
from google.genai.errors import APIError

# 2. Define the core logic function
def generate_response(api_key, user_input, prompt_mode):
    # Handle empty input validation
    if not api_key.strip():
        return "❌ Error: Please enter your Gemini API Key to run the application."
    if not user_input.strip():
        return "⚠️ Please enter some text or a question for the LLM to process."

    # Define the 3 system instructions based on selected mode
    prompt_templates = {
        "Explain Like I'm 5 🧸": (
            "You are an expert educator who excels at simplifying complex topics. "
            "Explain the user's input using simple language, fun analogies, and zero jargon. "
            "Imagine you are speaking to a 5-year-old child."
        ),
        "Professional Rewrite 💼": (
            "You are an elite corporate communications expert. Take the user's input "
            "and rewrite it to sound highly professional, polished, polite, and articulate. "
            "Maintain the original core intent but optimize for a business environment."
        ),
        "Generate Quiz Questions 🧠": (
            "You are a strict but fair professor. Based on the user's input text or topic, "
            "generate a high-quality quiz. Include 3 multiple-choice questions (with options A, B, C, D) "
            "and provide the correct answers hidden at the very bottom."
        )
    }

    system_instruction = prompt_templates.get(prompt_mode, "You are a helpful assistant.")

    try:
        # Initialize the new Google GenAI client with user's key
        client = genai.Client(api_key=api_key.strip())

        # Call the Gemini 2.5 Flash model with system instructions
        response = client.models.generate_content(
            model='gemini-2.5-flash',
            contents=user_input,
            config=types.GenerateContentConfig(
                system_instruction=system_instruction,
                temperature=0.7,
            )
        )
        return response.text

    except APIError as e:
        return f"❌ API Error: Please check if your API key is valid. Details: {e}"
    except Exception as e:
        return f"❌ An unexpected error occurred: {str(e)}"

# 3. Build the Gradio UI/UX Interface
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("# 🪄 Prompt Alchemist Application")
    gr.Markdown(
        "Experiment with how different prompt instructions dramatically alter an LLM's persona and output style. "
        "*Note: Your API key is processed locally and never stored.*"
    )

    with gr.Row():
        with gr.Column(scale=1):
            gr.Markdown("### ⚙️ Configuration")
            api_key_input = gr.Textbox(
                label="Step 1: Enter your Gemini API Key",
                placeholder="AIzaSy...",
                type="password"
            )

            prompt_mode_input = gr.Dropdown(
                label="Step 2: Choose Prompt Mode",
                choices=["Explain Like I'm 5 🧸", "Professional Rewrite 💼", "Generate Quiz Questions 🧠"],
                value="Explain Like I'm 5 🧸"
            )

            submit_btn = gr.Button("Transform Output ✨", variant="primary")

        with gr.Column(scale=2):
            gr.Markdown("### 📝 Workspace")
            user_text_input = gr.Textbox(
                label="Step 3: Enter your Topic or Text",
                placeholder="e.g., 'What is Quantum Computing?' or 'Hey, need this email sent to boss asap'",
                lines=4
            )

            output_display = gr.Markdown(label="Generated Response", value="*The LLM output will appear here...*")

    # Connect components to the backend function
    submit_btn.click(
        fn=generate_response,
        inputs=[api_key_input, user_text_input, prompt_mode_input],
        outputs=output_display
    )

# 4. Launch the application inside Colab
demo.launch(debug=True)
```
